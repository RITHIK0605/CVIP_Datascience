**RENAMING THE COLUMNS**

```python
df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'state',
                       'region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed',
                       'nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                       'weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
```

```python
df=df[['Year','Month','Day','Country','state','Region','city','AttackType','Killed','Wounded','Group']]
```

**NULL VALUES**

```python
df.isnull().sum()
```

**HANDLING NULL VALUES**

```python
df["Killed"]=df["Killed"].fillna('Unknown')
df["Wounded"]=df["Wounded"].fillna("Wounded")
```

**ABOUT THE MOST NUMBER OF ATTACKS**

```python
print("Country with the most attacks:",df['Country'].value_counts().idxmax())
print("City with the most attacks:",df['city'].value_counts().index[1])
print("Region with the most attacks:",df['Region'].value_counts().idxmax())
print("Year with the most attacks:",df['Year'].value_counts().idxmax())
print("Month with the most attacks:",df['Month'].value_counts().idxmax())
print("Group with the most attacks:",df['Group'].value_counts().index[1])
print("State with the most attacks:",df['state'].value_counts().idxmax())
```

**VISUALIZATION OF ATTACK OF YEARS**

```python
x_year = df['Year'].unique()
y_count_years = df['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,10))
sns.barplot(x = x_year,
           y = y_count_years)
plt.xticks(rotation = 45)
plt.xlabel('Attack Year')
plt.ylabel('Number of Attacks each year')
plt.title('Attack_of_Years')
plt.show()
```

**VISUALIZATION OF ATTACK OF MONTHS**

```python
x_month = df['Month'].unique()
y_count_month = df['Month'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,10))
sns.barplot(x = x_year,
           y = y_count_years)
plt.xticks(rotation = 45)
plt.xlabel('Attack Month')
plt.ylabel('Number of Attacks each month')
plt.title('Attack_of_Months')
plt.show()
```

**VISUALIZATION OF ATTACK OF DAYS**

```python
x_day = df['Day'].unique()
y_count_day = df['Day'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,10))
sns.barplot(x = x_year,
           y = y_count_years)
plt.xticks(rotation = 45)
plt.xlabel('Attack Day')
plt.ylabel('Number of Attacks each day')
plt.title('Attack_of_Days')
plt.show()
```

**VISUALIZATION OF TERRORIST ACTIVITIES BY REGION IN EACH YEAR**

```python
pd.crosstab(df.Year, df.Region).plot(kind='area',figsize=(15,6))
plt.title('Terrorist Activities by Region in each Year')
plt.ylabel('Number of Attacks')
plt.show()
```
